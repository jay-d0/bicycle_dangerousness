```python
with open('../data/HangJeongDong_ver20230101.geojson.txt', 'r', encoding='utf-8') as f:
  file = f.readlines()
  file
  f.close()
```

```python
concat_file = ''.join(file)
type(concat_file)
```

```python
map_dict = eval(concat_file)
map_dict.keys()
```

```python
map_dict['features'][0]['properties'].keys()
```

```python
sidonms = list()
for dong in map_dict['features']:
  sidonms.append(dong['properties']['sidonm'])
print(set(sidonms))
```

```python
new_features = list()
for dong in map_dict['features']:
  if dong['properties']['sidonm'] == '대전광역시':
    new_features.append(dong)
map_dict['features'] = new_features
```

```python
cnt = 0
for dong in map_dict['features']:
  print(dong['properties'])
  cnt += 1
print(cnt)
```

```python
with open('대전_HangJeongDong_ver20230101.geojson', 'w') as f:
  f.write(str(map_dict))
  f.close
```
